```python
import numpy as np
```

```python
# define data
x = np.array([ 1.,   1.5,  2.,   2.5,  3.,   3.5,  4.,   4.5,  5.,   5.5,  6.,   6.5,  7.,   7.5, 8.,   8.5,  9.,   9.5, 10. ])
y = np.array([3.43, 4.94, 6.45, 9.22, 6.32, 6.11, 4.63, 8.95, 7.8, 8.35, 11.45, 14.71, 11.97, 12.46, 17.42, 17.0, 15.45, 19.15, 20.86])
```

```python
#linear and cubic fit
linear = np.poly1d(np.polyfit(x,y,1))
cubic = np.poly1d(np.polyfit(x,y,3))
```

```python
#plotting
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(x,y)
plt.plot(x,linear(x))
plt.plot(x,cubic(x))
```

```python
#area under curve
from scipy.integrate import simps
```

```python
area_linear = simps(linear(x),x)
area_cubic = simps(cubic(x),x)
print(area_linear)
print(area_cubic)

plt.scatter(x,y)
plt.plot(x,linear(x))
plt.plot(x,cubic(x))

area1 = r"$\int_{x_0}^{x_{n-1}} linear(x)\mathrm{d}x = $" + "{:.0f}".format(area_linear)
area2 = r"$\int_{x_0}^{x_{n-1}} linear(x)\mathrm{d}x = $" + "{:.0f}".format(area_cubic)

plt.text(8, 7, area1,
         horizontalalignment='center', fontsize=15)
plt.text(8, 4, area2,
         horizontalalignment='center', fontsize=15)
plt.savefig("mlang_01.png")
```

```python
#Baysian information criterion
def BIC(y, yhat, k, weight = 1):
    err = y - yhat
    sigma = np.std(np.real(err))
    n = len(y)
    B = n*np.log(sigma**2) + weight*k*np.log(n)
    return B
```

```python
print(BIC(y,linear(x),2))
print(BIC(y,cubic(x),4))
```
